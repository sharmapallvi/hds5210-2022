### 1. Basic Statistics

Use Python to read in your data file and show many rows and columns your data has.

```python
import pandas as pd

nadacData = pd.read_csv("nadac2022.csv")

# The data file has 199120 rows and 12 columns
nadacData.shape
```

### 2. Data Diversity

Use Python to identify how many unique values each column in your data has.  Use Python to identify which column has the greatest number of distinct values (aka the most diversity)?  If your data has a unique identifier (e.g. a person or visit or record ID) then find the column with the next most number of unique values.

```python
# list of dataframe headers
columns = list(nadacData.columns.values)

# Keys are column headers and values are number of unique values
ColumnDistinct = {column: nadacData[column].nunique() for column in columns}

# Key-value pair with most diversity
max(ColumnDistinct.items(), key = lambda k : k[1])
```

### 3. Data Redundancy

Use Python to identify any columns in your data where the value for every row is the same.  For example, if you had a gender column and all the rows said 'M', then you would want to write the code that helped you determine that.

```python
SameValueColumns = []

for key, value in ColumnDistinct.items():
    # if value is 1, the value for every row in the column is the same
    if value == 1:
        SameValueColumns.append(key)

# Display columns where value of every row is the same
SameValueColumns
```

```python
groupedNadacData = nadacData.groupby(['As of Date', 'OTC'], as_index=False).mean()
pivotedNadacData = groupedNadacData.pivot(index='As of Date', columns='OTC', values='NADAC_Per_Unit')
```

```python
pivotedNadacData.plot.pie(subplots=True, figsize=(12, 6))
```

```python
pivotedNadacData.plot.bar(figsize=(18, 3))
```

The following observations can be made, based on the pie charts and bar graph:
1. Non-OTC products have a significantly higher average price per unit than OTC products.
2. There average price per unit for both OTC and non-OTC products remained relatively stable throughout.
